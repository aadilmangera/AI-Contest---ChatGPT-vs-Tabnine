### Response

To solve this problem, we can use a prefix sum array to keep track of the cumulative sum of haybales for each cow. We will also maintain a variable to keep track of the current difference in the number of haybales between Bessie and Elsie.

Here is the Python code to solve the problem:

```python
def fair_split_haybales():
    n = int(input())
    haybales = list(map(int, input().split()))
    q = int(input())

    # Calculate prefix sum for Bessie and Elsie
    prefix_sum = [0] * (n + 1)
    for i in range(n):
        prefix_sum[i + 1] = prefix_sum[i] + haybales[i]

    for _ in range(q):
        l, r, x = map(int, input().split())
        bessie_sum = prefix_sum[r] - prefix_sum[l - 1]
        ellie_sum = (r - l + 1) - bessie_sum
        difference = bessie_sum + x - ellie_sum
        print(difference)

fair_split_haybales()
```

This code reads the input, calculates the prefix sum for Bessie and Elsie, and then processes each query to calculate the difference in the number of haybales between Bessie and Elsie after splitting the given range of haybales. The difference is then printed for each query.

The time complexity of this solution is O(N + Q), where N is the number of haybales and Q is the number of queries. This is because we calculate the prefix sum in O(N) time and process each query in O(1) time.
