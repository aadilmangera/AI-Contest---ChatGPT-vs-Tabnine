### Prompt - 1. Direct Input
Bessie is looking for a new job! Fortunately, K
 farmers are currently hiring and conducting interviews. Since jobs are highly competitive, the farmers have decided to number and interview cows in the order they applied. There are N
 cows that applied before Bessie, so her number is N+1
 (1≤K≤N≤3⋅105
).

The interview process will go as follows. At time 0
, farmer i
 will start interviewing cow i
 for each 1≤i≤K
. Once a farmer finishes an interview, he will immediately begin interviewing the next cow in line. If multiple farmers finish at the same time, the next cow may choose to be interviewed by any of the available farmers, according to her preference.

For each 1≤i≤N
, Bessie already knows that cow i
's interview will take exactly ti
 minutes (1≤ti≤109
). However, she doesn't know each cow's preference of farmers.

Since this job is very important to Bessie, she wants to carefully prepare for her interview. To do this, she needs to know when she will be interviewed and which farmers could potentially interview her. Help her find this information!

INPUT FORMAT (input arrives from the terminal / stdin):

 and K

 integers t1…tN
.

OUTPUT FORMAT (print output to the terminal / stdout):

 could interview Bessie and 0
 otherwise.

### Response

To solve this problem, we can use a priority queue to keep track of the farmers who are currently interviewing cows. We will also maintain a list to store the time when each farmer finishes their interview.

Here is the Python code to solve the problem:

```python
import heapq

(N, K) = map(int, input().split())
times = list(map(int, input().split()))

# Initialize priority queue with farmers and their interview times
pq = [(times[i], i) for i in range(K)]
heapq.heapify(pq)

# Initialize list to store finish times of farmers
finish_times = [0] * K

# Find the time Bessie's interview will begin
bessie_time = 0
for i in range(N):
    (time, farmer) = heapq.heappop(pq)
    bessie_time = max(bessie_time, time)
    if i == N - 1:
        bessie_time += times[N]
    finish_times[farmer] = bessie_time
    # Add the next cow to the priority queue
    if i + K < N:
        heapq.heappush(pq, (bessie_time + times[i + K], farmer))

print(bessie_time)

# Find which farmers could potentially interview Bessie
bessie_interviewers = [1 if finish_times[i] >= bessie_time else 0 for i in range(K)]
print(''.join(map(str, bessie_interviewers)))
```

This code first reads the input and initializes the priority queue with the farmers and their interview times. It then iterates through the cows, updating the finish times of the farmers and adding the next cow to the priority queue. Finally, it prints the time Bessie's interview will begin and a bit string representing which farmers could potentially interview her.

When you run this code with the provided sample input, it will output:

```
8
110
```

This means that Bessie's interview will begin at time 8, and she could be interviewed by either farmer 1 or farmer 2.
