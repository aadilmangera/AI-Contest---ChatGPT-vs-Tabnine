### Response

To solve this problem, we can use a simple approach that involves calculating the number of times each fence post is touched during the daily walks of the cows. We can iterate through each cow and check if the starting and ending positions of the cow's walk are on the same side of a fence segment. If they are, we increment the touch count for the fence posts at the endpoints of that segment.

Here is the Python code that implements this approach:

```python
def count_touches(N, P, fence_posts, cows):
    touches = [0] * P

    for cow in cows:
        x1, y1, x2, y2 = cow

        for i in range(P):
            x3, y3 = fence_posts[i]
            x4, y4 = fence_posts[(i + 1) % P]

            # Check if the starting and ending positions are on the same side of the fence segment
            if (x3 - x1) * (y4 - y1) == (y3 - y1) * (x4 - x1) and (x3 - x2) * (y4 - y2) == (y3 - y2) * (x4 - x2):
                touches[i] += 1
                touches[(i + 1) % P] += 1
                break

    return touches

# Read input
N, P = map(int, input().split())
fence_posts = []
for _ in range(P):
    x, y = map(int, input().split())
    fence_posts.append((x, y))

cows = []
for _ in range(N):
    x1, y1, x2, y2 = map(int, input().split())
    cows.append((x1, y1, x2, y2))

# Calculate touches
touches = count_touches(N, P, fence_posts, cows)

# Print output
for touch in touches:
    print(touch)
```
