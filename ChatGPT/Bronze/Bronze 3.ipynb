```python
import sys
input = sys.stdin.read

def reconstruct_permutations(test_cases):
    results = []

    for N, h in test_cases:
        candidates = list(range(1, N + 1))
        p = [None] * N
        is_valid = True
        left, right = 0, N - 1

        for i in range(N - 2, -1, -1):
            if h[i] == candidates[left]:
                p[i + 1] = candidates[left]
                left += 1
            elif h[i] == candidates[right]:
                p[i + 1] = candidates[right]
                right -= 1
            else:
                is_valid = False
                break

        if is_valid:
            p[0] = candidates[left]
            results.append(" ".join(map(str, p)))
        else:
            results.append("-1")

    return results

# Input Parsing
data = input().split()
T = int(data[0])
index = 1
test_cases = []

for _ in range(T):
    N = int(data[index])
    h = list(map(int, data[index + 1:index + N]))
    test_cases.append((N, h))
    index += N

# Solve and Output
results = reconstruct_permutations(test_cases)
sys.stdout.write("\n".join(results) + "\n")
```
